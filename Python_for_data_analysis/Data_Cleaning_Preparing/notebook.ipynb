## Data Cleaning and Preparation
Before a machine learning model can find patterns in data, that data must be mathematically sound. Most ML algorithms (like those in Scikit-Learn) will throw a fatal error the moment they encounter a missing value.

In Pandas, missing data usually shows up as `NaN` (Not a Number) or `None`.

Filtering out (or dropping) missing data is the simplest way to handle this, but it must be done strategically. If you just delete everything that has a blank cell, you might throw away 80% of your valuable dataset.

#### Filling in Missing values

In Pandas, the workhorese for this is the .fillna() method. The strategy you choose depends entirely on the type of data in that column.

**1. The Constant Fill ( Best for Text/categorical Data)**

if  a non-crirtical text column is missing, we simply replace the NaN with a standardized placeholder. This prevents our ML models from crashing while explicitly marking the data as missing.

```python
df
```

```python
# let's delete the Carrier_ID , which is NaN , since there's no use of it.
cleaned_df = df.dropna(subset=['Carrier_ID'])
cleaned_df
```

#### Removing the Duplicates

In the world of data, duplicates are dangerous. If you are training a machine learning model to predict shipping costs, feeding it the exact same company twice gives that company unfair weight in the algorithm's "memory."

**Step 1: The Diagonals (`duplicated()`)**
Before we delete anything, we should see what Pandas considers a duplicate. By default, Pandas only flags a row as a duplicate if **every single columns** matches exactly.


```python
# Returns True for any row that is an exact copy of a previous row
cleaned_df.duplicated().sum()
```

**Step 2: Targeted Duplication (The `subset` Parameter)**

Just like we did with missing data, we have to tell Pandas to focus only on the critical columns. If two rows share the exact same `Carrier_ID`, it is the same company, regardless of what the `Notes` column says

```python
# Check for duplicates based ONLY ON Carrier_ID
print(cleaned_df.duplicated(subset=['Carrier_ID']).sum())
```

**Step 3: The Execution (`drop_duplicates()`)**

To actually remove the rows, we use `.drop_duplicates()`. But when you have two versions of the same row, you have to decide which one survives. We control this using the keep parameter.

we have three strategic choices:
- `keep='first'` **(The default):** Keeps the very first instance it sees reading top-to-bottom and deletes all subsequent copies.
- `keep='last` : Deletes the first instances and keeps the very last copy at the bottom of the file (useful if you know newer data is appended to the bottom).
- `keep=False` **(The Nuclear Options)**: If it finds a duplicate, it deletes all versions of it. Neither survives. (Useful in fraud detection where any duplication is highly suspicious).

```python
# Drop duplicates based on the ID, keeping the first occurence, and overwriting the variable
cleaned_df = cleaned_df.drop_duplicates(subset=['Carrier_ID'], keep='first')
print(cleaned_df)
```

```python
import pandas as pd
import numpy as np

# 1. Load the data into a BRAND NEW variable name (test_df)
test_df = pd.read_csv('messy_registry_data.csv', index_col=False, skipinitialspace=True, usecols=range(8))

# 2. Let's look at the raw data BEFORE we touch it
print("--- 1. RAW TEXT DATES ---")
print(test_df['Registration_Date'].dropna().head())

# 3. Now apply the conversion
test_df['Registration_Date'] = pd.to_datetime(test_df['Registration_Date'], format='mixed', errors='coerce')

# 4. Let's look at the data AFTER the conversion
print("\n--- 2. AFTER CONVERSION ---")
print(test_df[['Carrier_ID', 'Registration_Date']].head())
```

```python
# Remove the NaN vlaue on Carrier_ID
cleaned_test_df = test_df.dropna(subset=['Carrier_ID']).copy()
# cleaned_test_df

# Remove the duplicates
cleaned_test_df = cleaned_test_df.drop_duplicates(subset=['Carrier_ID'], keep='first')
cleaned_test_df
```

```python
import re

def clean_fleet(value):
    if pd.isna(value):
        return pd.NA
    val_str = str(value).lower().strip()
    # Dictionary to map common text numbers to digits
    word_to_num = {
        'one' : '1', 'two': '2', 'three': '3', 'four': '4', 'five': '5',
        'six': '6', 'seven': '7', 'eight': '8', 'nine': '9', 'ten': '10',
        'eleven': '11', 'twelve': '12'
    }
    
    # If the exact string matches a key in our dictionary, replace it with
    # the digit version
    if val_str in word_to_num:
        val_str = word_to_num[val_str]
    
    # The Regex Magic: Replace any non-digit character (\D) with nothing
    digits_only = re.sub(r'\D', '', val_str)

    # If stripping letters left us with an empty string, return missing
    if digits_only == '':
        return pd.NA
    
    # Convert the final clean string into an integer
    return int(digits_only)

# 2. Apply the function to the column
cleaned_test_df['Fleet_Size'] = cleaned_test_df['Fleet_Size'].apply(clean_fleet)

# 3. Convert the 'Fleet_Size' into Int64.
cleaned_test_df['Fleet_Size'] = cleaned_test_df['Fleet_Size'].astype('Int64')

cleaned_test_df
```
